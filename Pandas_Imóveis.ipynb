# **Manipulando dados para o time de Dev**

```python
#Trabalhando com o arquivo original e completo
dados_dev = pd.read_csv('/content/aluguel.txt', sep=';')
dados_dev.head()
```

```python
#Criando coluna Valor_Mes com a soma de Aluguel e Condominio
dados_dev['Valor_Mes'] = dados_dev['Valor'] + dados_dev['Condominio']
dados_dev.head()
```

```python
#Criando Coluna Valor_Ano considerando Aluguel, condominio e IPTU dos 12 meses
dados_dev['Valor_Ano'] = dados_dev['Valor_Mes'] * 12 + dados_dev['IPTU']
dados_dev.head()
```

```python
# Criar colunas categóricas
# Coluna Descrição usando concatenação de strings
# No caso da info não ser string precisa ser editada
# A \ serve para trocar de linha sem interferir no codigo
dados_dev['Descricao'] = dados_dev['Tipo'] + ' em ' + dados_dev['Bairro'] + ' com ' + \
dados_dev['Quartos'].astype(str) + ' quarto(s) e ' + dados_dev['Vagas'].astype(str) + \
' vaga(s) de garagem.'
dados_dev.head()
```

```python
# Criar coluna binária (Boolean) para suites utilizando funcao lambda
dados_dev['Possui Suite'] = dados_dev['Suites'].apply(lambda x: "Sim" if x > 0 else "Nao")
dados_dev.head()
```

```python
dados_dev.to_csv('dados_dev.csv', sep=';', index=False)
```

***1) Os alunos participaram de uma atividade extracurricular e ganharam pontos extras. Esses pontos extras correspondem a 40% da nota atual de cada um deles. Com base nisso, crie uma coluna chamada "Pontos_extras" que contenha os pontos extras de cada aluno, ou seja, 40% da nota atual deles.***

```python
alunos['Pontos Extras'] = alunos['Notas'] * 0.4
alunos
```

***2) Crie mais uma coluna, chamada "Notas_finais" que possua as notas de cada aluno somada com os pontos extras.***

```python
alunos['Notas Finais'] = (alunos['Notas'] + alunos['Pontos Extras'])
alunos['Notas Finais'] = alunos['Notas Finais'].apply(lambda x: 10 if x > 10 else x)
alunos

```

***3) Como houve uma pontuação extra, alguns alunos que não tinham sido aprovados antes podem ter sido aprovados agora. Com base nisso, crie uma coluna chamada "Aprovado_final" com os seguintes valores:***

***True: caso o aluno esteja aprovado (nota final deve ser maior ou igual a 6); False: caso o aluno esteja reprovado (nota final deve ser menor que 6).***

```python
alunos['Aprovado_Final'] = alunos['Notas Finais'].apply(lambda x: 'Aprovado' if x >= 6 else 'Reprovado')
alunos
```
